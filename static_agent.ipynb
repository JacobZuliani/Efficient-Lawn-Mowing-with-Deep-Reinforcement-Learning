```python
class mower():

    def __init__(self, environment_state, position, direction):
        # environment_state is a 2d lawn array the mower will be acting in
        # position is a list containing a coordinate pair [x,y]
        # direction is one of the four characters N, S, E, W
        self.history = environment_state
        self.position = position
        self.last_position = position
        self.last_block = environment_state[self.position[0], self.position[1]]
        self.direction = direction
        self.moves = 0
        __update_history()
    
    def __update_history(self)
        # updates the mowers history with the last move, must be run after every action
        
        # place mower in its current location
        self.history[self.position[0], self.position[1]] = 4
        # replace block last occupied by mower
        self.history[self.last_position[0], self.last_position[1]] = self.last_block
    
    def advance(self):
        self.last_position = self.position
        if self.direction == 'N':
            self.position[1] += 1
        elif self.direction == 'S':
            self.position[1] -= 1
        elif self.direction == 'E':
            self.position[0] += 1
        elif self.direction == 'W':
            self.position[0] -+ 1
        else:
            raise Exception('unknown direction')
        self.moves += 1
         __update_history(self)
    
    def pivot_clockwise(self):
        if self.direction == 'N':
            self.direction = 'E'
        elif self.direction == 'S':
            self.direction = 'W'
        elif self.direction == 'E':
            self.direction = 'S'
        elif self.direction == 'W':
            self.direction = 'N'
        else:
            raise Exception('unknown direction')
        self.moves += 1
        
    def pivit_counterclockwise(self):
        if self.direction == 'N':
            self.direction = 'W'
        elif self.direction == 'S':
            self.direction = 'E'
        elif self.direction == 'E':
            self.direction = 'N'
        elif self.direction == 'W':
            self.direction = 'S'
        else:
            raise Exception('unknown direction')
        self.moves += 1
        
    def get_position():
        return self.x, self.y
    
    def get_direction():
        return self.direction
```
