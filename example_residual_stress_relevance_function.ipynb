```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def example_residual_stress_relevance_function(residual_stress):
    if residual_stress < 1:
        return 0
    elif residual_stress > 6:
        return 10
    else:
        return (np.e**(1.3*residual_stress))/((np.e**residual_stress)+1)
```

```python
plt.figure(figsize=(8,6))
plt.plot([example_residual_stress_importance_function(residual_stress) for residual_stress in np.arange(0, 10, 0.01)], color = 'blue', linewidth=2.5)
plt.hlines(6, 0, 1000, linestyles = 'dashed', label='Scrap Threshold', color = 'red', linewidth=3)
plt.hlines(1, 0, 1000, linestyles = 'dashed', label='Irrelevance Threshold', color = 'green', linewidth=3)
plt.title('Example Residual Stress Relevance Function', fontsize = '15')
plt.xlabel('Relevance of Residual Stress', fontsize = '15')
plt.ylabel('Residual Stress', fontsize = '15')
plt.legend()
plt.grid()
plt.show()
```
